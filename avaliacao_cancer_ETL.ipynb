```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
import tabula
```

```python
# Chamar a função read_pdf para extrair a tabela
list_pdf = tabula.read_pdf("cancer_observatory.pdf", pages=[19]) #o argumento pages all extrairá todas as tabelas, caso necessário aplicar pages=[1,2,3]

# Converter a lista de DataFrames em um único DataFrame
df_pdf = pd.concat(list_pdf)

# Visualizar o DataFrame
df_pdf
```

```python
# Remover os parênteses de cada valor em cada linha de cada coluna
for column in df_pdf.columns:
    df_pdf[column] = df_pdf[column].str.replace(r'\(.*\)', '', regex=True)

# Visualizar o DataFrame
df_pdf.head(5)
```
